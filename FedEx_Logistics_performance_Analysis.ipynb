## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. Handling Missing Values for Categorical Data
# Filling gaps in 'Shipment Mode' with the most frequent value (Mode)
df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])

# 2. Cleaning Numeric Columns
# Converting text errors to NaN, then filling with the Median to maintain data distribution
numeric_cols = ['Weight (Kilograms)', 'Freight Cost (USD)', 'Line Item Insurance (USD)']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].median())

# 3. Date Formatting
# Converting strings into Datetime objects for time-based calculations
date_cols = ['Scheduled Delivery Date', 'Delivered to Client Date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 4. Feature Engineering: Creating 'Delivery_Delay'
# Metric: Actual Delivery Date - Scheduled Delivery Date (Positive = Late, Negative = Early)
df['Delivery_Delay'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# 5. Feature Engineering: Extracting Year for Trend Analysis
df['Arrival_Year'] = df['Delivered to Client Date'].dt.year

print("Data Wrangling Complete!")
print(f"New KPI 'Delivery_Delay' created for storytelling.")
df[['Weight (Kilograms)', 'Freight Cost (USD)', 'Delivery_Delay']].head()
# 6. String Standardization (Crucial for Charts 1, 2, and 9)
# Removes hidden spaces and makes naming consistent
df['Shipment Mode'] = df['Shipment Mode'].str.strip().str.title()
df['Country'] = df['Country'].str.strip()

# 7. Quality Control: Drop rows where Delay couldn't be calculated
# If we don't have both dates, we can't measure performance.
df = df.dropna(subset=['Delivery_Delay'])

# 8. Filter Outliers for Financial Accuracy
# Removing shipments with 0 or Negative weight/cost which are data entry errors
df = df[(df['Weight (Kilograms)'] > 0) & (df['Freight Cost (USD)'] > 0)]

print("Final Polish Complete!")
print(f"Cleaned Dataset Shape: {df.shape}")
````
